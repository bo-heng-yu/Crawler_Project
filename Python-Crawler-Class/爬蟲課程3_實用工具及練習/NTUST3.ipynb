## 動手實作

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://feebee.com.tw/product/iPhone%20X%20256G%20%E6%99%BA%E6%85%A7%E5%9E%8B%E6%89%8B%E6%A9%9F/?q=iphone%20x')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
price = []
for iPhone_price in soup.find_all(class_ = 'price ellipsis'):
    p = int(iPhone_price.text.replace(',', ''))
    price.append(p)
    print(p)
```

```python
sum(price)/len(price)
```

## 蘋果日報實戰

```python
url = 'https://tw.appledaily.com/hot/daily'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
for i in soup.find(class_ = 'all').find_all('li'):
    num = i.find(class_ = 'aht_title_num').text
    title = i.find(class_ = 'aht_title').text
    view = i.find(class_ = 'aht_pv_num').text
    print(num, title, view)
```

## 除了文字，還想將每一則的網址搜集下來怎麼做？

```python
for i in soup.find(class_ = 'all').find_all('li'):
    num = i.find(class_ = 'aht_title_num').text
    title = i.find(class_ = 'aht_title').text
    view = i.find(class_ = 'aht_pv_num').text
    url =  i.find(class_ = 'aht_title').find('a').get('href')
    print(num, title, view, url)
```

## 動手實作

```python
url = 'http://news.ltn.com.tw/list/newspaper'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
for i in soup.find(class_='list').find_all('li'):
    title = i.find(class_ = 'tit').find('p').text
    tag = i.find(class_ = 'tagarea').text
    url = i.find(class_ = 'tit').get('href')
    print([title, tag, url])
```
